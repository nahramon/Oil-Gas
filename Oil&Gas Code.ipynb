Transformation

```python
# Delete rows not going to be used
df = df.drop(columns = {'prod_agua', 'iny_agua', 'iny_gas', 'iny_co2', 'iny_otro', 'vida_util', 'observaciones', 'fechaingreso', 'rectificado', 'habilitado',
'sigla', 'vida_util','sub_tipo_recurso', 'clasificacion', 'subclasificacion'})
```

```python
# Check for null or None values in cuenca column
null_cuenca = df['cuenca'].isnull()

# Get rows with null or None values
rows_with_null = df[null_cuenca]
rows_with_null

# 36 Null rows in 'cuenca' but those have 'tipoestado' = Abandonado which means that it is no longer in use.

# Check for null or None values in tipoextraccion column
null_cuenca = df['tipoextraccion'].isnull()

# Get rows with null or None values
rows_with_null = df[null_cuenca]
rows_with_null

# 41 Null rows by 'tipoextraccion' but those have no production at all.
# Same with 'tipoestado' and 'tipopozo'
```

```python
# Rename columns: prod_pet and prod_gas since its units are different

df = df.rename(columns={'prod_pet': 'prod_pet_m3', 'prod_gas': 'prod_gas_Mm3'})
```

```python
# Convert 'date_column' to datetime
df['fecha_data_time'] = pd.to_datetime(df['fecha_data'])

# Extract month and year, and format them as '%Y-%m' (year-month)
df['fecha_data_time'] = df['fecha_data_time'].dt.to_period('M')

df
```

```python
grouped_df = df.groupby(['fecha_data_time'])['prod_pet_m3'].sum()
print(grouped_df)
```

```python
# Delete rows with Nan values int the column "tipoextraccion" since they are garbage data
df.dropna(subset=['tipoextraccion'], inplace=True)
df.info
```

```python
# Filter rows where "tipo_de_recurso" is equal to "SIN RESERVORIO" and remove them
df = df[df['tipo_de_recurso'] != 'SIN RESERVORIO']
df
```
